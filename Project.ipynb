```python
import pandas as pd
df=pd.read_csv(r"C:/Users/marag/Documents/GitProject/Pakistan_Largest_Ecommerce_Dataset.csv")
```

## Data Cleaning

```python
df.head()
```

```python
df=df.drop(['Unnamed: 21','Unnamed: 22','Unnamed: 23','Unnamed: 24','Unnamed: 25'], axis =1)
```

```python
df.dropna(subset='item_id',inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df['item_id']=df['item_id'].astype(str)
df['created_at']=pd.to_datetime(df['created_at'])
df= df.rename(columns={'created_at':'order_date'})
df['status']=df['status'].replace(to_replace=['order_refunded','refund'],value='refunded')
df['status']=df['status'].replace(to_replace=['paid','received'],value='complete')
df['status']=df['status'].replace(to_replace=['pending_paypal','processing','holded'],value='pending')
```

```python
df['Customer ID']=df['Customer ID'].astype(str)
```

```python
df['order_date'].min()
```

## Top 5 Best Selling Category

```python
pd.set_option('float_format','{:f}'.format)
```

```python
Best_selling_cat=df[df.status=='complete'].groupby('category_name_1').agg({'Customer ID':'nunique','grand_total':'sum'}).sort_values(by='grand_total',ascending=False).reset_index()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
```

```python
Best_selling_cat.head(5).plot.bar(x='category_name_1',y='grand_total')
plt.title('Best Selling Categories')
plt.xlabel('Category')
plt.ylabel('Total Amount')
plt.xticks(rotation=45)
plt.show()
```

## Average basket per category

```python
df[df.status=='complete'].groupby('category_name_1')['grand_total'].mean().sort_values(ascending=False).plot(kind='bar')
```
